```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from PIL import Image
from pathlib import Path
```

```python
image_dir_train="./data/m_train2/m_train/images"
csv_file_train="./data/m_train2/m_train/train.csv"

image_dir_valid="./data/m_valid/m_valid/images"
csv_file_valid="./data/m_valid/m_valid/valid.csv"

image_dir_test = "./data/m_test/m_test/images"
```

```python
file_count_train = len(list(Path(image_dir_train).glob("*.tif")))
file_count_valid = len(list(Path(image_dir_valid).glob("*.tif")))
file_count_test = len(list(Path(image_dir_test).glob("*.tif")))

file_counts = {'train': [file_count_train],
               'val': [file_count_valid],
               'test': [file_count_test]}

df_counts = pd.DataFrame(file_counts).transpose()
df_counts
```

```python
plt.figure(figsize=(6,6))
plt.pie(df_counts[0], labels=df_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightsalmon', 'lightgreen'])
plt.show()
```

```python
df = pd.read_csv(csv_file_train)
df_proportion = pd.DataFrame(df['cls'].value_counts())
df_proportion.columns = ['count']
df_proportion['count'].to_dict()
```

```python
plt.figure(figsize=(6,6))
plt.pie(df_proportion['count'], labels=df_proportion.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightsalmon', 'lightgreen'])
plt.show()
```
